### Estructura de las Clases

Definición de Clases

Animal<br>
>|------>Perro <br>
>>|------>Labrador<br>


>|------>Gato<br>
>>|------>GatoPersa

```python
class Animal():
  def __init__(self, Nombre):
    self.nombre=Nombre
    self.comida=0
  
  def comer (self):
    self.comida+=1


class Perro(Animal):
  
  def __init__(self, Nombre):
    super().__init__(Nombre)
    self.ladridos=0
  
  def ladrar(self):
    self.ladridos+=1

class Labrador(Perro):
  
  def __init__(self, Nombre):
    super().__init__(Nombre)
    self.raza='labrador'
  
class Gato(Animal):
  
  def __init__(self, Nombre):
    super().__init__(Nombre)
    self.maullidos=0
 
  def maullar(self):
    self.maullidos+=1

class GatoPersa(Gato):
  
  def __init__(self, Nombre):
    super().__init__(Nombre)
    self.raza='persa'    
```

```python
#Clases que heredan de Animal
Animal.__subclasses__()
```

```python
#Clases que heredan de Perro
Perro.__subclasses__()
```

```python
#Clases que heredan de Gato
Gato.__subclasses__()
```

```python
#Defino una función para listar todas las subclases
def get_all_subclasses(cls):
    all_subclasses = []

    for subclass in cls.__subclasses__():
        all_subclasses.append(subclass)
        all_subclasses.extend(get_all_subclasses(subclass))

    return all_subclasses
```

```python
#Listo todas las subclases de Animal
get_all_subclasses(Animal)
```

### Ejemplo herencia múltiple

#### Definimos dos superclases 

```python
class SuperClase1():
  def __init__(self):
    self.nombre1='Superclase1'
    self.padre1='clase1'

class SuperClase2():
  def __init__(self):
    self.nombre2='Superclas2'
    self.padre2='clase2'
```

##### Creamos la subclase heredando de ambas superclases.

Usamos super() en el constructor

```python
class SubClase(SuperClase1, SuperClase2):
  def __init__(self):
    super().__init__()
    self.hijo='Subclase'
```

```python
#Creamos el objeto e imprimimos sus atributos
objeto=SubClase()
print(objeto.hijo)
print(f'padre1: {objeto.padre1}')
print(f'padre2: {objeto.padre2}')
```

super() hace referencia a la primera superclase dentro de la definicion de la subclase  *class subClase( **SuperClase1**, SuperClase2)* por lo tanto no se considero el constructor de la superclase2 y el atributo padre2 no se creó.

##### Probemos lo mismo invirtiendo el orden de las superclases:

```python
class SubClase(SuperClase2, SuperClase1):
  def __init__(self):
    super().__init__()
    self.hijo='Subclase'
```

```python
#Creamos el objeto e imprimimos sus atributos
objeto=SubClase()
print(objeto.hijo)
print(f'padre1: {objeto.padre1}')
print(f'padre2: {objeto.padre2}')
```

En este caso no se creó el atributo padre1

##### Probemos llamando a los constructores de ambas superclases

```python
class SubClase(SuperClase2, SuperClase1):
  def __init__(self):
    SuperClase2.__init__(self)
    SuperClase1.__init__(self)
    self.hijo='Subclase'
```

```python
#Creamos el objeto e imprimimos sus atributos
objeto=SubClase()
print(objeto.hijo)
print(f'padre1: {objeto.padre1}')
print(f'padre2: {objeto.padre2}')
```

Se crearon todos los atributos heredados

##### Probemos una forma más genérica de llamar a los constructores de ambas superclases

```python
class SubClase(SuperClase2, SuperClase1):
    def __init__(self):
        for clase in self.__class__.__bases__:
            clase.__init__(self)
        
        self.hijo='Subclase'
```

```python
#Creamos el objeto e imprimimos sus atributos
objeto=SubClase()
print(objeto.hijo)
print(f'padre1: {objeto.padre1}')
print(f'padre2: {objeto.padre2}')
```
