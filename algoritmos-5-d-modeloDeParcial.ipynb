#Ejercicio 1
a) Escribir una función, usando el método iterativo, que dada una lista de números, devuelva la misma lista pero donde a los números impares se les sume +1. Por ejemplo, dado
[2,5,4,3] -> [2,6,4,4]
[3,3,3] -> [4,4,4]
[] -> []
[4] -> [4]  

```python
def incrementar_impares(lista):
  for i in range(len(lista)):
   es_impar = (lista[i]%2 == 1) # me da true si el elemento que estoy viendo es impar
   if (es_impar):
     lista[i] += 1
     # lista[i] = lista[i] + 1
  return lista
```

```python
l1 = [2, 5, 4, 3]
l2 = [3, 3, 3]
l3 = []
l4 = [4]

print(incrementar_impares(l1))
print(incrementar_impares(l2))
print(incrementar_impares(l3))
print(incrementar_impares(l4))
```

# Ejercicio 2
Implementar una clase llamada Pava_electrica. La clase tiene como atributos: cantidad_liquido, temperatura, fabricante y linea. Además, tiene los métodos: __init__, cargar_liquido, vaciar_liquido y calentar_liquido.

```python
class Pava_electrica:
  def __init__(self, cantidad_liquido, temperatura, fabricante, linea):
    self.cantidad_liquido = cantidad_liquido
    self.temperatura = temperatura
    self.fabricante = fabricante
    self.linea = linea

  def cargar_liquido(self, cantidad):
    self.cantidad_liquido += cantidad

  # El enunciado no dice cuánto se quiere vaciar asi que asumo que la persona va a elegir la cantidad a vaciar.
  def vaciar_liquido(self, cantidad):
    self.cantidad_liquido = self.cantidad_liquido - cantidad

  def calentar_liquido(self, temperatura_deseada):
    if self.cantidad_liquido > 0:
      if temperatura_deseada > self.temperatura:
        print(f"Calentando el liquido de {self.temperatura} a {temperatura_deseada}")
        self.temperatura = temperatura_deseada
      else:
          print("La temperatura deseada es mayor que la temperatura actual.")
    else:
      print("No hay liquido para calentar.")
```

```python
pava_motorola = Pava_electrica(200, 20, "motorola", "nose")
```

# Ejercicio 3

Implemente una función recursiva que reciba dos números positivos y enteros como parámetros $(n, \ p)$ y  devuelva el resultado de ejecutar  la siguiente operación:
$F(n, p) = (1 * p) + (2 * p) + (3 * p) +  … + ( n * p)$

Por ej, si llamamos F(2, 2), debe devolver 6. Si llamamos $F(3,2)$ debe devolver 12.


```python
# F(2,2) = 2 * 2 + 1 * 2 + 0 * 2

# F(3,2) = 3 * 2 + 2 * 2 + 1 * 2 + 0 * 2

# F(n,p) = n*p + F(n-1, p)

def suma_recursiva(n,p):
  if n==0:
    return 0
  else:
    return n*p + suma_recursiva(n-1,p)
```

```python
suma_recursiva(2,2)
```

```python
suma_recursiva(3,2)
```

Dados los siguientes programas:

```python
def s1(lista): # O(n) donde n es el tamaño de la lista
  j, i = 0, len(lista)
  while j < i:
      a,b = lista[j],lista[i-1]
      j = j + 1
      i = i - 1
  return a == b, i == j
```

```python
def s2(lista): # O(n) + O(n) = O(n)
    z = 0
    j = len(lista)-1   # tamaño lista - 1
    # recorro de atrás para adelante y guardo en z la cantidad de elementos pares
    while j >= 0:
      if lista[j]%2==0:  # el elemento de la lista es par
        z = z + 1
      j = j - 1
    # j = -1
    # recorro de adelante para atrás y guardo en z la cantidad de elementos (pares - impares)
    while j < len(lista)-1:
      if lista[j+1]%2!=0: # el elemento de lista es impar
        z = z - 1
      j = j + 1
    return z < 0, z

    # z < 0 : hay mas impares que pares? o hay la misma cantidad?
    # z : cantidad de pares - cantidad de impares
```

```python
s2([2,3,1,12])
```

```python
s2([2,3,2,2,3,2])
```

```python
def s3(lista): # O(n)
    z = 0
    i = 0
    while i < len(lista):
        if(lista[i] % 2 == 0): # el elemento es par
            z = z+1 # sumamos uno si es par
        else:
            z = z-1 # sumamos uno si es impar
        i = i+1
    return z<0, z
# z < 0 : tengo más impares que pares? o igual cantidad?
# z : cantidad pares - cantidad de impares
```
