### Selection Sort

```python
#Ordena una lista de elementos según el método de selección.
#Retorna la lista ordenada

def selection_sort(lista):
  i = 0
  while (i < len(lista)):
    posmin = i
    j = i+1
    while (j < len(lista)):
      if (lista[j] < lista[posmin]):
        posmin = j
      j = j+1
    
    aux = lista[posmin]
    lista[posmin] = lista[i]
    lista[i] = aux
    i = i+1
  #  print('i: ',i,'lista: ',lista)
#  return lista
```

```python
#Busca el menor elemento de una lista entre la posición a y b
#Retorna su posición

def buscar_min(lista, a, b):
  
  pos_min = a
 
  for i in range(a + 1, b + 1):
    if lista[i] < lista[pos_min]:
     pos_min=i
  
  return pos_min
 
```

```python
#Ordena una lista de elementos según el método de selección.
#Retorna la lista ordenada

def selection_sort(lista):
  
  # posición inicial de la lista
  n = 0

  # mientras haya al menos 2 elementos para ordenar
  while n < len(lista):
    
    # posición del menor valor del segmento de la lista
    pos_min = buscar_min(lista, n, len(lista)-1 )

    # intercambiar el valor que está en pos_min con el valor que
    # está en la primera posición del segmento de la lista
    lista[pos_min], lista[n] = lista[n], lista[pos_min]
    
    # reducir el segmento en 1
    n = n + 1
  
  return lista 
```

```python
selection_sort([10,9,8,7,6,5])
```

###Insertion Sort

#### Una implementación

```python
# Ordena una lista de elementos según el método de inserción.
# Retorna la lista ordenada

def insertion_sort(lista):  
  i = 0
  while (i < len(lista)):   
   
    elem = lista[i] #elemento de la lista a ordenar
    j=i-1

    while (j>=0 and elem < lista[j]):
      lista[j+1] = lista[j]  #mueve los mayores hacia la derecha 
      j = j-1
    
    lista[j+1] = elem   # inserta el elemento

    i = i+1  #avanza i
  
  return lista
```

```python
insertion_sort([59,7,388,41,2,280,50,123])
```

#### Otra implementación

```python
# Reubica al elemento que está en la posición p de la lista dentro del segmento [0:p-1].
def reubicar(lista, p):
  
  #elementoa reubicar
  elem = lista[p]

  # Recorrer el segmento [0:p-1] de derecha a izquierda hasta
  # encontrar la posición j tal que lista[j-1] <= v < lista[j].
  j = p
  while j > 0 and elem < lista[j - 1]:
    # Desplazar los elementos hacia la derecha, dejando lugar
    # para insertar el elemento elem donde corresponda.
    lista[j] = lista[j - 1]
    j -= 1

  lista[j] = elem
```

```python
# Ordena una lista de elementos según el método de inserción.
# Retorna la lista ordenada

def insertion_sort(lista):
  
  for i in range(len(lista) - 1):
    # Si el elemento de la posición i+1 está desordenado respecto
    # al de la posición i, reubicarlo dentro del segmento [0:i]
    if lista[i + 1] < lista[i]:
       reubicar(lista, i + 1)

  return lista
```

### Bubble Sort

```python
# Ordena una lista de elementos según el método de burbujeo.
# Retorna la lista ordenada

def bubble_sort(lista):
  
  i = 0
  
  while (i < len(lista)):
    
    j = 0 
    while (j < len(lista)-1):
      
      if (lista[j]>lista[j+1]): # avanza los elementos mayores
        aux = lista[j+1]
        lista[j+1] = lista[j]
        lista[j] = aux
     
      j = j+1
    
    i = i+1
    
  return lista
```

```python
bubble_sort([59,7,388,41,2,280,50,123])
```

## Recursión Algoritmica

```python
### producto iterativo
def prod(m,n):
  RV = 0
  while m > 0 :
    RV = RV + n
    m = m - 1

  return RV
```

```python
prod(5,4)
```

```python
### producto recursivo
def prod_rec(m,n):
  if m == 0:
    RV = 0
  else:
    RV = prod_rec(m-1, n) + n
```

```python
### Factorial iterativo
def fact(n):
  RV = 1
  while n > 0:
    RV = RV * n
    n = n-1
  return RV
```

```python
### Factorial recursivo
def fact_rec(n):
  if n == 0:
    RV = 1
  else:
    RV = n * fact_rec(n-1)
  return RV
```

### Mergesort Recursivo

#### Una implementación

```python
def merge_sort(A):
  if len(A)>1:
    A_izq = A[:len(A)//2]
    A_der = A[len(A)//2:]

    #recursion
    merge_sort(A_izq)
    merge_sort(A_der)

    #merge
    i = 0 # indice A_izq
    j = 0 # indice A_der
    k = 0 #indice de la lista combinada
    while i < len(A_izq) and j < len(A_der):
      if A_izq[i] < A_der[j]:
        A[k] = A_izq[i]
        i = i + 1
      else:
        A[k] = A_der[j]
        j = j + 1
      k = k + 1

    while i < len(A_izq):
      A[k] = A_izq[i]
      i = i + 1
      k = k + 1
    while j < len(A_der):
      A[k] = A_der[j]
      j = j + 1
      k = k + 1
  return A
```

```python
lista = [5, 6, 8, 6, 7, 9, 47, 1, 8, 6, 4, 6, 5]

print(merge_sort(lista))
```

#### Otra implementación

```python
#Intercala los elementos de lista1 y lista2 de forma ordenada.
#Pre: lista1 y lista2 deben estar ordenadas.
#Devuelve: una lista con los elementos de lista1 y lista2.

def merge(lista1, lista2):
  
  i, j = 0, 0
  resultado = []

 #Intercala ordenado menor a mayor 
  while(i < len(lista1) and j < len(lista2)):
    if (lista1[i] < lista2[j]):
      resultado.append(lista1[i])
      i += 1
    else:
      resultado.append(lista2[j])
      j += 1

 # Agregar lo que falta
  resultado += lista1[i:]
  resultado += lista2[j:]

  return resultado
```

```python
#Ordena una lista mediante el método merge sort.
#Pre: la lista debe contener elementos comparables.
#Devuelve: una nueva lista ordenada.

def merge_sort(lista):

  if len(lista) < 2:
    return lista
  
  medio = len(lista) // 2
  
  #recursion
  izq = merge_sort(lista[:medio])
  der = merge_sort(lista[medio:])
 
  return merge(izq, der) #merge
```

```python
lista = [5, 6, 8, 6, 7, 9, 47, 1, 8, 6, 4, 6, 5]

print(merge_sort(lista))
```

MCD.PNG

```python
#Resolución

def mcd(a,b):
  if b==0:
    return a
  if a==0:
    return b
  
  if a>=b:
     return mcd(a-b,b)
  else:
     return mcd(a,b-a)

print(mcd(180,32))
```

### Ejercicio 5

Implementar un algorítmo de busqueda binaria recursiva

```python
# Resolución

def buscar(x,A):
  esta, pos = False, -1
  if len(A) != 0:
    esta, pos = BuscarDesdeHasta(x, A, 0, len(A)-1)

  return esta, pos

def BuscarDesdeHasta(x, A, desde, hasta):
  if desde == hasta:
    if x == A[desde]:
      esta, pos = True, desde
    else:
      esta, pos = False, -1
  else:
    med = (desde + hasta)//2
    if A[med] < x:
      esta, pos = BuscarDesdeHasta(x, A, med + 1, hasta)
    else:
      esta, pos = BuscarDesdeHasta(x, A, desde, med)
  return esta, pos

lista=[1,2,3,4,5,6,7,8]
buscar(7,lista)
```
