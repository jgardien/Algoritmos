#Ejercicio 1
a) Escribir una función, usando el método iterativo, que dada una lista de números, devuelva la misma lista pero donde a los números impares se les sume +1. Por ejemplo, dado
[2,5,4,3] -> [2,6,4,4]
[3,3,3] -> [4,4,4]
[] -> []
[4] -> [4]  

```python
def incrementar_impares(lista):
    for i in range(len(lista)):
        es_impar = lista[i]%2 == 1 # vale True si el elemento es impar
        if (es_impar):
            lista[i] += 1
    return lista
```

```python
l1 = [2,5,4,3]
l2 = [3,3,3]
l3 = []
print(incrementar_impares(l1))
print(incrementar_impares(l2))
print(incrementar_impares(l3))
```

# Ejercicio 2
Implementar una clase llamada Pava_electrica. La clase tiene como atributos: cantidad_liquido, temperatura, fabricante y linea. Además, tiene los métodos: __init__, cargar_liquido, vaciar_liquido y calentar_liquido.

```python
class Pava_electrica:
    def __init__(self, cantidad_liquido, temperatura, fabricante, linea):
        self.cantidad_liquido = cantidad_liquido
        self.temperatura = temperatura
        self.fabricante = fabricante
        self.linea = linea

    def cargar_liquido(self, cantidad):
            self.cantidad_liquido += cantidad


    def vaciar_liquido(self):
        self.cantidad_liquido = 0


    def calentar_liquido(self, temperatura_deseada):
        if self.cantidad_liquido > 0:
            if temperatura_deseada > self.temperatura:
                print(f"Calentando el líquido de {self.temperatura}°C a {temperatura_deseada}°C")
                self.temperatura = temperatura_deseada
            else:
                print("La temperatura deseada debe ser mayor que la temperatura actual.")
        else:
            print("No hay líquido para calentar.")
```

```python
pava_dia = Pava_electrica(200, 20, 'dia', 'pe')
```

```python
pava_dia.cantidad_liquido
```

```python
pava_dia.cargar_liquido(100)
```

```python
pava_dia.cantidad_liquido
```

```python
pava_dia.calentar_liquido(40)
```

# Ejercicio 3

Implemente una función recursiva que reciba dos números positivos y enteros como parámetros $(n, \ p)$ y  devuelva el resultado de ejecutar  la siguiente operación:
$F(n, p) = (1 * p) + (2 * p) + (3 * p) +  … + ( n * p)$

Por ej, si llamamos F(2, 2), debe devolver 6. Si llamamos $F(3,2)$ debe devolver 12.


```python
def suma_recursiva(n,p):
    if n==0:
       return 0
    else: return n*p + suma_recursiva(n-1, p)
```

```python
suma_recursiva(3,2)
```

# Ejercicio 4

```python
def s1(lista):
  j, i = 0, len(lista)
  while j < i:
      a,b = lista[j],lista[i-1]
      j = j + 1
      i = i - 1
  return a == b
```

```python
def s2(lista):
    z = 0
    j = len(lista)-1
    while j >= 0:
      if lista[j]%2==0:
        z = z + 1
      j = j - 1
    while j < len(lista)-1:
      if lista[j+1]%2!=0:
        z = z - 1
      j = j + 1
    return z < 0
```

```python
l7 = [2,2,2,2,2]

print(s2(l1))
print(s2(l2))
print(s2(l3))
print(s2(l4))
print(s2(l5))
print(s2(l6))
print(s2(l7))
```
